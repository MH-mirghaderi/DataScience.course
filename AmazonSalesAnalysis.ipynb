```python
import pandas as pd

cols = ['discounted_price', 'actual_price', 'discount_percentage','rating_count','rating']

def clean_and_convert(s):
    return (
        s.astype(str)
         # ۱. حذف فاصله‌های غیرمعمول (از جمله ZWNJ)
         .str.replace(r'[\u200c\s]', '', regex=True)
         # ۲. حذف جداکننده‌های هزارگان انگلیسی و فارسی
         .str.replace(r'[٬,]', '', regex=True)
         # ۳. حذف همه‌چیز به‌جز ارقام و نقطه
         .str.replace(r'[^0-9\.]', '', regex=True)
         # ۴. تبدیل نهایی به عدد (خطاها => NaN)
         .pipe(pd.to_numeric, errors='coerce')
    )

for col in cols:
    df[col] = clean_and_convert(df[col])
```

```python
df=df.dropna(subset=['rating_count','rating'])
```

#1 Statistical Questions


##1. Does the discounted price significantly impact product ratings?

```python
from scipy.stats import spearmanr
rho , pval = spearmanr(df['discounted_price'] , df['rating'])
print(f'spearmans ρ = {rho:.4f} , p-value = {pval:.4f}' )
```

### Result

>

1.   ρ (rho):
                this is spearman's rank-correlation coeficient.
  its value ranges from -1 to +1:


*   +1 means the ranks move perfectly together in the same (increasing) direction (perfect positive monotonic assosiation)
*   -1 means the ranks move perfectly in opposite direction(perfect negative monotonic assosiation)
*   0 means there's no clear rank-order relationship.

2.   p-value


*   this is the probability that, under the null hypothesis of 'no associaton' between the variables
*   if the p-value is below our chosen significance level(commonly 0.05), we reject the null and conclude that the observed correlation is statistically signicant.

Result : ρ = 0.0801 indicates a very,very weak association.
According to points that denoted above , actually no real association.






##2. Are product categories and high/low ratings independent?

```python
from scipy.stats import chisquare,chi2_contingency
```

```python
df['rating-level'] = np.where(df['rating'] >=4, 'High', 'Low')
```

```python
df['Main_Category'] = df['category'].str.split('|').str[0]
```

```python
contingency = pd.crosstab(df['Main_Category'], df['rating-level'],margins = True)
display(contingency)
```

```python
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"Chi2 statistic: {chi2:.2f}")
print(f"p-value: {p:.3e}")
print(f"Degrees of freedom: {dof}")
```



###Result
Since p < 0.05, we reject the null hypothesis of independence.
There is a statistically significant relationship between product category and whether a product receives a high (≥ 4) or low (< 4) rating.



##3. Is there a significant difference in ratings between high-discount and lowdiscount products?


```python
df['discount_group'] = df['discount_percentage'].apply(lambda x: 'High' if x > 50 else 'Low')
high = df[df['discount_group'] == 'High']['rating']
low = df[df['discount_group'] == 'Low']['rating']
tstat, pval = stats.ttest_ind(high, low, equal_var=False)
print("T-statistic:", tstat)
print("P-value:", pval)
```

###Result
The average score between the two groups differs significantly. Products with high and low discounts are not comparable in terms of scores, and this difference is significant.

##4. Do more expensive products receive higher ratings on average?

```python
df['price_group'] = pd.qcut(df['actual_price'], 3, labels=['Low', 'Medium', 'High'])
groups = [group['rating'].dropna() for name, group in df.groupby('price_group')]
anova_stat, pval = stats.f_oneway(*groups)
print("ANOVA F-statistic:", anova_stat)
print("P-value:", pval)
```

###Result
There is no significant difference in scores among the price groups (low, medium, high). Therefore, it cannot be said that more expensive products necessarily receive higher scores.

##5. Does the distribution of rating counts follow a normal distribution?

```python
counts = df['rating_count'].dropna()
ks_stat, pval = stats.kstest((counts - counts.mean()) / counts.std(), 'norm')
print("K-S Statistic:", ks_stat)
print("P-value:", pval)
```

###Result
The number of scores does not follow a normal distribution, and this difference is statistically highly significant. Therefore, nonparametric tests should be used for this variable.
